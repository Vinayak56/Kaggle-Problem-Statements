```python
df1 = pd.read_csv("./test.csv")
```

```python
df1.shape
```

```python
del df1["Cabin"]
del df1["Name"]
del df1["Ticket"]
```

```python
df1["Age"].fillna(value=df1["Age"].mean(),inplace=True)
df1 = pd.get_dummies(df1,columns=["Pclass","Sex","SibSp","Parch","Embarked"],drop_first=True)
```

```python
oulier1 = df1["Age"].quantile(0.99)
oulier2 = df1["Age"].quantile(0.01)
df1.Age[df1["Age"]>1.005*oulier1] = 1.005*oulier1
df1.Age[df1["Age"]<0.8*oulier2] = 0.8*oulier2
```

```python
del df1["Parch_4"]
del df1["Parch_5"]
del df1["Parch_6"]
del df1["Parch_3"]
del df1["SibSp_2"]
del df1["SibSp_3"]
del df1["SibSp_4"]
del df1["SibSp_5"]
del df1["SibSp_8"]
del df1["PassengerId"]
outlier1 = df1["Fare"].quantile(0.99)
df1.Fare[df1.Fare>1.05*outlier1] = 1.05*outlier1
```

```python
del df1["Parch_9"]
```

```python
df1.head()
```

```python
Xmm = df1.loc[:,df1.columns!="Survived"]
Ymm = xgb_clf.predict(Xmm)
```

```python
Xmm.shape
```

```python
df2 = pd.read_csv("./gender_submission.csv")
from sklearn.metrics import accuracy_score
accuracy_score(df2["Survived"],Ymm)
```

```python
from sklearn.model_selection import GridSearchCV
xgb_clff = xgb.XGBClassifier(n_estimators=400,learning_rate=0.1,random_state=42)
param_test1 = {
    "max_depth" : range(3,10,2),
    "gamma" : [0.1,0.2,0.3],                
    "subsample" : [0.8,0.9],            
    "colsample_bytree" : [0.8,0.9],          
    "reg_alpha" : [1e-2,0.1,1]                 
}
```

```python
grid_search = GridSearchCV(xgb_clff,param_test1,cv=5,scoring="accuracy",n_jobs=-1)
```

```python
grid_search.fit(Xm,Ym)
```

```python
grid_search.best_params_
```

```python
bst_mdl = grid_search.best_estimator_
```

```python
accuracy_score(df2["Survived"],bst_mdl.predict(Xmm))
```
